```python
import pandas as pd
import json
import mysql.connector
```

```python

#SE USO UN ARCHIVO JSON OBTENIDO DE https://www.ecdc.europa.eu/en/publications-data/data-virus-variants-covid-19-eueea

with open('datos.json') as archivo:
    datos = pd.read_json(archivo)

#print(datos)
```

```python
# Eliminar todas las filas con datos faltantes
datos_limpio = datos.dropna()

# Eliminar todas las columnas con datos faltantes
datos_limpio = datos.dropna(axis=1)

#print(datos_limpio)
df_filtrado = datos_limpio.iloc[:1000, [0, 1, 2, 6, 7]]

print(df_filtrado)
```

```python
mycursor = mydb.cursor()

mycursor.execute("USE covid19")

mycursor.execute("CREATE TABLE datos(country VARCHAR(255), year_week  VARCHAR(255), source  VARCHAR(255), variant VARCHAR(255) , number_detections_variant VARCHAR(255))")

```
